```python
import pandas as pd

gesch = pd.read_csv('./data/TSV/Geschichte-30.tsv', sep='\t')
gesch['type'] = 'gesch'
print('len gesch: {}'.format(len(gesch)))

gesell = pd.read_csv('./data/TSV/Gesellschaft-30.tsv', sep='\t')
gesell['type'] = 'gesell'
print('len gesell: {}'.format(len(gesell)))
```

```python
all_data = pd.concat([gesch, gesell], ignore_index=True)

all_data.drop_duplicates(subset='pageid', inplace=True, keep=False)

len(all_data)
```

```python
all_data.drop('number', axis=1, inplace=True)
all_data.drop('namespace', axis=1, inplace=True)

all_data.reset_index(drop=True, inplace=True)

all_data
```

```python
all_data.to_csv('./data/all_data.csv', index=False)
```
