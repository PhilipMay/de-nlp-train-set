```python
import pandas as pd

gesch_test = pd.read_csv('./data/gesch_test.csv')
gesch_train = pd.read_csv('./data/gesch_train.csv')
gesell_test = pd.read_csv('./data/gesell_test.csv')
gesell_train = pd.read_csv('./data/gesell_train.csv')

gesch_test_ids = gesch_test['pageid'].tolist()
gesch_train_ids = gesch_train['pageid'].tolist()
gesell_test_ids = gesell_test['pageid'].tolist()
gesell_train_ids = gesell_train['pageid'].tolist()
```

```python
articles_saved_count = 0

def is_doc_start_line(line):
    return line.startswith('<doc id="')

def is_doc_end_line(line):
    return line.startswith('</doc>')

def read_id(line):
    index_1 = line.index('"')
    index_2 = line.index('"', index_1 + 1)
    return line[index_1+1:index_2]

def get_id_category(doc_id):
    result = None
    
    if doc_id in gesch_test_ids:
        result = 'gesch_test'

    elif doc_id in gesch_train_ids:
        result = 'gesch_train'

    elif doc_id in gesell_test_ids:
        result = 'gesell_test'

    elif doc_id in gesell_train_ids:
        result = 'gesell_train'
        
    return result
        
        
def save_content(content, doc_id, category):    
    global articles_saved_count
    striped_content = []
    
    for line in content:
        if line.strip() != '':
            striped_content.append(line)
    
    with open('data/' + category + '/' + str(doc_id) + '.txt', 'x') as the_file:
        for line in striped_content:
            the_file.write(line + '\n')

    articles_saved_count += 1

    if articles_saved_count % 1000 == 0:
        print('Saved id: {} category: {} articles_saved_count: {}'.format(doc_id, category, articles_saved_count))
```

```python
read_content_area = False
content = []

with open('../wikiextractor/2/wiki_00', "r") as input_file:
    for line in input_file:

        if is_doc_start_line(line):
            doc_id = int(read_id(line))
            category = get_id_category(doc_id)
            if category != None:
                read_content_area = True
                
        elif is_doc_end_line(line):
            if read_content_area:
                save_content(content, doc_id, category)
            read_content_area = False
            content = []

        elif read_content_area:
            line = line.replace('\n', '')
            content.append(line)
```
