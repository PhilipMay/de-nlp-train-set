```python
import pandas as pd

pages = pd.read_csv('./data/all_data.csv')

pages['available'] = False

page_ids = pages['pageid'].tolist()
```

```python
def is_doc_start_line(line):
    return line.startswith('<doc id="')

def read_id(line):
    index_1 = line.index('"')
    index_2 = line.index('"', index_1 + 1)
    return line[index_1+1:index_2]

def read_this_id(doc_id):
    return doc_id in page_ids
```

```python
read_content_area = False
content = []

ids_found_count = 0

with open('../wikiextractor/2/wiki_00', "r") as input_file:
    for line in input_file:

        if is_doc_start_line(line):
            doc_id = int(read_id(line))
            if read_this_id(doc_id):
                pages.loc[pages['pageid'] == doc_id, 'available'] = True
                ids_found_count += 1
                if ids_found_count % 10000 == 0:
                    print('Written id: {}, ids_found_count: {}'.format(doc_id, ids_found_count))
```

```python
pages.to_csv('./data/all_available_data.csv', index=False)
```
