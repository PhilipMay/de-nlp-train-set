```python
import pandas as pd
```

```python
all_available_data = pd.read_csv('./data/all_available_data.csv')

all_available_data
```

```python
gesch = all_available_data[all_available_data['type'] == 'gesch']
av_gesch = gesch[gesch['available'] == True]
av_gesch = av_gesch.sample(frac=1, random_state=42)
av_gesch
```

```python
gesell = all_available_data[all_available_data['type'] == 'gesell']
av_gesell = gesell[gesell['available'] == True]
av_gesell = av_gesell.sample(frac=1, random_state=42)
av_gesell = av_gesell.iloc[:len(av_gesch)]
av_gesell
```

```python
print('len av_gesch: {}'.format(len(av_gesch)))
print('len av_gesell: {}'.format(len(av_gesell)))
```

```python
gesch_test = av_gesch.iloc[:30000]
gesch_train = av_gesch.iloc[30000:]

gesell_test = av_gesell.iloc[:30000]
gesell_train = av_gesell.iloc[30000:]
```

```python
print('len gesch_test: {}'.format(len(gesch_test)))
print('len gesch_train: {}'.format(len(gesch_train)))
print('len gesell_test: {}'.format(len(gesell_test)))
print('len gesell_train: {}'.format(len(gesell_train)))
```

```python
gesch_test.to_csv('./data/gesch_test.csv', index=False)
gesch_train.to_csv('./data/gesch_train.csv', index=False)
gesell_test.to_csv('./data/gesell_test.csv', index=False)
gesell_train.to_csv('./data/gesell_train.csv', index=False)
```
